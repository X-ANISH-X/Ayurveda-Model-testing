```python
import pandas as pd

dataset = pd.read_excel("AYUR_VAIDYA_DATASET_generated_10000_entries.xlsx")
dataset.head()
```

```python
dataset.columns
```

```python
features = ['AGE', 'GENDER', 'COMORBIDITIES',
       'SPECIFIC ALLERGIES', 'BAD HABITS', 'SYMPTOM 1', 'SYMPTOM 2',
       'SYMPTOM 3', 'SYMPTOM 4', 'PROB 1', 'PROB 2', 'DISEASE', 'VAATA', 'PITTA', 'KAPHA']
categorical = ['GENDER', 'COMORBIDITIES',
       'SPECIFIC ALLERGIES', 'BAD HABITS', 'SYMPTOM 1', 'SYMPTOM 2',
       'SYMPTOM 3', 'SYMPTOM 4', 'PROB 1', 'PROB 2']
df = dataset[features]
df.loc[:, :] = df.fillna("None")
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

df = df[features] # Can't directly use df cause of the first col

# Using new LabelEncoder for each col and storing it for later conversion back to text


train, test = train_test_split(df, test_size = 0.2, random_state=69)

label_encoders = {col: LabelEncoder() for col in categorical+['DISEASE']}
for col in label_encoders:
    train[col] = label_encoders[col].fit_transform(train[col])

X_train = train[['AGE', 'GENDER', 'COMORBIDITIES',
       'SPECIFIC ALLERGIES', 'BAD HABITS', 'SYMPTOM 1', 'SYMPTOM 2',
       'SYMPTOM 3', 'SYMPTOM 4', 'PROB 1', 'PROB 2', 'VAATA', 'PITTA', 'KAPHA']]
Y_train = train['DISEASE']
X_test = test[['AGE', 'GENDER', 'COMORBIDITIES',
       'SPECIFIC ALLERGIES', 'BAD HABITS', 'SYMPTOM 1', 'SYMPTOM 2',
       'SYMPTOM 3', 'SYMPTOM 4', 'PROB 1', 'PROB 2', 'VAATA', 'PITTA', 'KAPHA']]
Y_test = test['DISEASE']

Y_test
```

```python
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=69)
xgb_model.fit(X_train, Y_train)
```

```python
for col in label_encoders:
    if col != 'DISEASE':
        X_test[col] = label_encoders[col].transform(X_test[col]) 
```

```python
test_pred = xgb_model.predict(X_test)

# Convert back to text 
test_pred_text = label_encoders['DISEASE'].inverse_transform(test_pred)
accuracy = accuracy_score(Y_test, test_pred_text)
print("Accuracy :", accuracy)
print(test_pred_text)
print(Y_test)
```
